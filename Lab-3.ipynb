```python
# Freqency domain using fourier transform
violin_ft = np.fft.fft(violin_c4)
violin_ft
```

```python
def plot_magnitude_spectrum(signal, sr, title, f_ratio=1):
    X = np.fft.fft(signal)
    X_mag = np.absolute(X)
    
    plt.figure(figsize=(5,3))
    
    f = np.linspace(0, sr, len(X_mag))
    f_bins = int(len(X_mag)*f_ratio)  
    
    plt.plot(f[:f_bins], X_mag[:f_bins])
    plt.xlabel('Frequency (Hz)')
    plt.title(title)
```

```python
plot_magnitude_spectrum(violin_c4, sr, "violin", 0.2)
```

```python
plot_magnitude_spectrum(piano_c5, sr, "Piano", 0.2)
```

```python
plot_magnitude_spectrum(woodwind_c4, sr, "WoodWind", 0.2)
```

```python
plot_magnitude_spectrum(noise, sr, "Noise", 0.2)
```

```python
signal, sr = librosa.load(piano_file)
signal.shape
```

```python
# extract mfcc 
mfcc = librosa.feature.mfcc(signal, n_mfcc = 7, sr = sr)
mfcc.shape
```

```python
# Visualize MFCCs
plt.figure(figsize = (8,5))
librosa.display.specshow(mfcc, x_axis = 'time', sr=sr)
plt.colorbar(format="%+2f")
plt.show()
```
